```python
#Google's Stock Price Analysis
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mp
import seaborn as s
```

```python
df=pd.read_csv("google.csv")
```

```python
df['Date']=pd.to_datetime(df['Date'])
df.info()
```

```python
df=df.set_index(df['Date'])
```

```python
mp.figure(figsize=(10,5))
vol=s.lineplot(data=df,x='Date',y='Volume')
```

```python
df.plot(y=['High','Close','Open'],x='Date', title="Google Stock Price")
```

```python
fig,(a1,a2,a3)=mp.subplots(3,figsize=(15,20))
df.groupby(df.index.day).mean().plot(y='Volume',ax=a1,xlabel='Day')
df.groupby(df.index.month).mean().plot(y='Volume',ax=a2,xlabel='Day')
df.groupby(df.index.year).mean().plot(y='Volume',ax=a3,xlabel='Day')
```

```python
#DATES WITH HIGHEST STOCK PRICE
```

```python
a= df.sort_values(by = 'High', ascending= False).head(5)
print(a['High'])
```

```python
#DATES WITH LOWEST STOCK PRICE
```

```python
a= df.sort_values(by = 'Low', ascending= True).head(5)
print(a['Low'])
```

```python
fig,axes=mp.subplots(nrows=1,ncols=2,figsize=(10,8))
fig.suptitle('High and Low values stock per period of time', fontsize=15)
s.lineplot(ax=axes[0],y=df['High'],x=df.index,color='green')
s.lineplot(ax=axes[1],y=df['Low'],x=df.index,color='red')
```
